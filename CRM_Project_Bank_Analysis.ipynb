```python
sns.set(style="whitegrid")
sns.barplot(x='Geography',y='Exited', hue='Gender',data=data)
```

```python
number_of_prod=pd.get_dummies(data['NumOfProducts'],prefix='NOP')
```

```python
# age-group of people who exited from the bank 
a=sns.FacetGrid(data,hue='Exited',aspect=6)
a.map(sns.kdeplot,'Age',shade=True)
a.set(xlim=[0,data['Age'].max()])
a.add_legend()
```

```python
age_bin = [18,22,34,40,60,80,100]
data_label = ['18-22','23-34','35-40','41-60','61-80','81-100']
age_labels=pd.cut(data['Age'],age_bin,labels=data_label,right=False)
```

```python
data['Age_labeled']= age_labels
```

```python
##Exited on the basis of No of credit cards
sns.barplot(x='Age_labeled',y='NumOfProducts',hue='Gender', data = data )
```

```python
## Exited on the basis of age 
sns.barplot(x = 'Age_labeled',y='Exited',hue='Gender',data=data,color="darkblue")
```

```python
#creating a group as per the FICO recommedtation
data_bins =[300,579,669,739,799,850]
data_labels =['Very Poor','Fair','Good','Very Good','Exceptional']
crdscore_labels=pd.cut(data['CreditScore'],data_bins,labels=data_labels)
data['CreditScore_labled']=crdscore_labels
creditscore = pd.get_dummies(data['CreditScore_labled'], prefix = 'CreditLevel')

```

```python
#Plotting graph
sns.barplot(x = 'CreditScore_labled',y='Exited',hue='Gender',data=data)
```

```python
#Splitting data according to tenure
data_bins = [0,1,5,8,11]
data_labels = ['0-1','1-5','5-8','8-11']
tenure_labels=pd.cut(data['Tenure'],data_bins,labels=data_labels,right=False)
data['Tenure_labeled']= tenure_labels
tenure = pd.get_dummies(data['Tenure_labeled'],prefix = 'Tenure')
sns.barplot(x = 'Tenure_labeled',y='Exited',hue='Gender',data=data)
```

```python
ext_data =data[['Exited', 'Age', 'NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']]
ext_data_corrs = ext_data.corr()
plt.figure(figsize = (6,6))
sns.heatmap(ext_data_corrs, cmap = plt.cm.RdYlBu_r, vmin = -0.6, annot = True, vmax = 0.6)
plt.title('Correlation Heatmap')
```
